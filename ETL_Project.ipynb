```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
crime_file = "crimebystatecombinedwithunemployment.csv"
crime_df = pd.read_csv(crime_file)
crime_df.head()
```

```python
wage_file = "Minimum Wage Data.csv"
wage_df = pd.read_csv(wage_file, encoding="cp1252")
wage_df.head()
```

```python
state_file = "state_abbrev.csv"
state_df = pd.read_csv(state_file)
state_df.drop(columns=["Unnamed: 2"], inplace=True)
state_df.head()
```

```python
crime_state = crime_df.merge(state_df, how="outer", left_on='state', right_on="Abbreviation")
crime_state.head()
                      
```

```python
crime_state.drop(columns=["state", "Abbreviation"], inplace=True)
```

```python
crime_state.reset_index(inplace=True)
crime_state.rename(columns={"index":"id","Population": "population", "Murder": "murder", "Robbery": "robbery", 
                          "Aggravated assault": "aggravated_assault", "Burglary":"burglary", 
                            "Larceny theft": "larceny_theft", "State":"state", "violent total": "violent_total",
                           "property total": "property_total", "vehicle theft": "vehicle_theft"}, inplace=True)

crime_state.head()
```

```python
crime_state_df = crime_state[["id", "year", "state", "population", "unemployment", "violent_total", "murder", "rape",
                             "robbery", "aggravated_assault", "property_total", "burglary", "larceny_theft",
                             "vehicle_theft"]].copy()

crime_state_df
```

```python
max_year_crime = crime_state_df["year"].max()
min_year_crime = crime_state_df["year"].min()
```

```python
wage_df.reset_index(inplace=True)
wage_df.rename(columns={"index": "id", "State": "state", "Year": "year", "State.Minimum.Wage": "state_min_wage", 
                        "State.Minimum.Wage.2020.Dollars": "state_min_wage_2020_dollars", "Federal.Minimum.Wage":"federal_min_wage", 
                        "Federal.Minimum.Wage.2020.Dollars": "federal_min_wage_2020_dollars",
                        "Effective.Minimum.Wage":"effective_min_wage", "Effective.Minimum.Wage.2020.Dollars": "effective_min_wage_2020_dollars",
                        "CPI.Average": "cpi_avg", }, inplace=True)
wage_df.drop(columns=["Department.Of.Labor.Uncleaned.Data", "Department.Of.Labor.Cleaned.Low.Value", "Department.Of.Labor.Cleaned.Low.Value.2020.Dollars",
             "Department.Of.Labor.Cleaned.High.Value", "Department.Of.Labor.Cleaned.High.Value.2020.Dollars", "Footnote"], inplace=True)


wage_df
```

```python
max_year_wage = wage_df["year"].max()
min_year_wage = wage_df["year"].min()
```

```python
new_wage = wage_df.loc[(wage_df['year']<=max_year_crime)| (wage_df['year']>=min_year_crime)]
new_wage
```

```python
new_wage = wage_df.loc[(wage_df['year']<=max_year_crime) & (wage_df['year']>=min_year_crime)]
new_wage
```
